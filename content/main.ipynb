**3. Imbalance Handling**

````python
import pandas as pd
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE

# Load the dataset
financial_distress_data = pd.read_csv('/content/Financial Distress.csv')

# Discretize the Financial Distress variable
bins = [-float('inf'), -2, -1, 0, 1, 2, float('inf')]
labels = ['Very Low', 'Low', 'Neutral', 'High', 'Very High', 'Critical']
financial_distress_data['Distress Category'] = pd.cut(financial_distress_data['Financial Distress'], bins=bins, labels=labels)

# Check the new distribution of the discretized variable
distress_category_distribution = financial_distress_data['Distress Category'].value_counts()

plt.figure(figsize=(8, 6))
distress_category_distribution.plot(kind='bar')
plt.title('Distribution of Discretized Financial Distress')
plt.xlabel('Distress Category')
plt.ylabel('Frequency')
plt.show()

# Separating the target variable and features
X = financial_distress_data.drop(columns=['Financial Distress', 'Distress Category'])
y = financial_distress_data['Distress Category']

# Applying SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Checking the new distribution after SMOTE
new_distress_category_distribution = pd.Series(y_resampled).value_counts()

plt.figure(figsize=(8, 6))
new_distress_category_distribution.plot(kind='bar')
plt.title('New Distribution of Discretized Financial Distress After SMOTE')
plt.xlabel('Distress Category')
plt.ylabel('Frequency')
plt.show()
````

**Imbalance Handling**

Initially, the `Financial Distress` variable exhibited class imbalance, which can lead to biased model predictions. To address this, the Synthetic Minority Over-sampling Technique (SMOTE) was applied to balance the dataset by generating synthetic samples for underrepresented classes.

**Distribution of Discretized Financial Distress**

The initial distribution of the discretized `Financial Distress` variable was highly imbalanced, with a majority of instances falling into certain distress categories. The application of SMOTE balanced the distribution, ensuring that each category is equally represented, which is crucial for training robust and unbiased predictive models.

**3. Additional Data Quality Check**

````python
# Additional Data Quality Check
# Check for duplicates
duplicates = credit_risk_data.duplicated().sum()
print("Number of duplicate rows:", duplicates)

# Identify outliers using IQR for numerical columns
def find_outliers(data):
    outliers = {}
    for col in data.select_dtypes(include=[np.number]).columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = ((data[col] < lower_bound) | (data[col] > upper_bound)).sum()
    return outliers

outliers = find_outliers(credit_risk_data)
print("Outliers in each numerical column:\n", outliers)

# Convert categorical columns to category dtype
credit_risk_data[categorical_cols] = credit_risk_data[categorical_cols].astype('category')

# Summary of data types and memory usage after processing
print("Data Types and Memory Usage:\n", credit_risk_data.info())
````

The dataset was cleaned and enhanced by creating a new feature, `Debt_to_Income_Ratio`, and normalizing the relevant columns. Outliers were identified, and categorical columns were optimized for memory usage.

**4. Graph-Based Techniques**

````python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes for each unique credit card
credit_cards = transactions['credit_card'].unique()
G.add_nodes_from(credit_cards)

# Add edges for each transaction (assuming transactions between different credit cards)
for index, row in transactions.iterrows():
    G.add_edge(row['credit_card'], row['credit_card'], weight=row['transaction_dollar_amount'])

# Ensure that the graph is not empty before visualization
if len(G.edges) > 0:
    # Visualize the graph
    plt.figure(figsize=(12, 12))
    pos = nx.spring_layout(G, k=0.1)
    nx.draw_networkx_nodes(G, pos, node_size=10)
    nx.draw_networkx_edges(G, pos, alpha=0.1)
    plt.title('Transaction Network Graph')
    plt.show()
else:
    print("The graph has no edges to display.")

# Calculate centrality measures
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

# Display top 5 nodes by degree centrality
top_degree_centrality = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 nodes by degree centrality:", top_degree_centrality)

# Identify anomalies based on centrality measures
# For instance, nodes with unusually high betweenness centrality might indicate fraudulent behavior
anomalies = {node: centrality for node, centrality in betweenness_centrality.items() if centrality > 0.01}
print("Anomalous nodes based on betweenness centrality:", anomalies)
````

**Creating Lag Features**

Before handling outliers or missing values, I'll create the `Lag_Financial_Distress` feature, which captures the previous period's financial distress value for each company. This feature is crucial for temporal modeling and can help predict future distress based on past values.

````python
import pandas as pd

# Reload the dataset
financial_distress_data = pd.read_csv('/content/Financial Distress.csv')

# Creating lag features for Financial Distress
financial_distress_data['Lag_Financial_Distress'] = financial_distress_data.groupby('Company')['Financial Distress'].shift(1)

# Check for any missing values introduced by lag feature
missing_values = financial_distress_data.isnull().sum()
print("Missing Values after creating Lag_Financial_Distress:\n", missing_values[missing_values > 0])
````


**Handling Missing Values**

Now that the `Lag_Financial_Distress` column has been created, I’ll handle the missing values that were introduced by the lagging process. These will be filled with the median of the `Financial Distress` column to maintain the integrity of the time series.

````python
# Fill missing values in Lag_Financial_Distress with the median value of Financial Distress
financial_distress_data['Lag_Financial_Distress'].fillna(financial_distress_data['Financial Distress'].median(), inplace=True)

# Verify that there are no more missing values in Lag_Financial_Distress
missing_values = financial_distress_data.isnull().sum()
print("Missing Values after handling Lag_Financial_Distress:\n", missing_values[missing_values > 0])
````



**Handling Outliers**

With the dataset fully prepared, I will now remove outliers using the Interquartile Range (IQR) method to ensure the remaining data is clean and suitable for modeling.

````python
import numpy as np

# Identify numerical columns
numerical_columns = financial_distress_data.select_dtypes(include=[np.number]).columns

# Function to remove outliers using the IQR method
def remove_outliers(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter data to remove outliers
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data

# Remove outliers
cleaned_financial_distress_data = remove_outliers(financial_distress_data, numerical_columns)

# Verify the number of rows after outlier removal
print(f"Number of rows after outlier removal: {cleaned_financial_distress_data.shape[0]}")
````

### **3.3.2 Predictive Analytics for Compliance Issues**

In this section, I developed and validated multiple machine learning models to predict potential compliance risks using the Financial Distress dataset. The objective was to identify the most effective model for forecasting compliance-related issues, a critical component in maintaining regulatory adherence within financial institutions.

#### **Data Preprocessing and Feature Engineering**

The Financial Distress dataset was first subjected to feature engineering to enhance its predictive capability:

- **Lagged_Financial_Distress**: This feature captures the financial distress level of the previous period for each company. By incorporating this temporal context, the model can leverage historical trends to predict future compliance risks.
- **Debt_to_Equity**: This ratio, calculated as the ratio of `x1` (a financial metric) to `x2` (another financial metric), is a key indicator of a company’s financial leverage. Higher leverage may indicate increased financial distress, thus representing a potential compliance risk.

After feature engineering, the dataset was cleaned by removing rows with missing values. The target variable was defined as a binary indicator, where a value of `1` denotes financial distress (indicating a compliance risk), and `0` denotes no distress.

To address the issue of class imbalance—a common problem in financial distress datasets—I applied the Synthetic Minority Over-sampling Technique (SMOTE). This technique generated synthetic samples for the minority class, ensuring the model is not biased towards the majority class.


````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# Load the Financial Distress dataset
financial_distress_data = pd.read_csv('Financial Distress.csv')

# Feature Engineering
financial_distress_data['Lagged_Financial_Distress'] = financial_distress_data.groupby('Company')['Financial Distress'].shift(1)
financial_distress_data['Debt_to_Equity'] = financial_distress_data['x1'] / (financial_distress_data['x2'] + 1e-5)

# Drop rows with missing values
financial_distress_data.dropna(inplace=True)

# Create a binary target variable: 1 for distress (compliance risk), 0 for no distress
financial_distress_data['Target'] = (financial_distress_data['Financial Distress'] < -1).astype(int)

# Select relevant features
X = financial_distress_data[['Lagged_Financial_Distress', 'Debt_to_Equity', 'x1', 'x2', 'x3']]
y = financial_distress_data['Target']

# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````



#### **Model Training**

Four different machine learning models were trained to predict compliance risks:

1. **Random Forest Classifier**: This ensemble model aggregates the predictions of multiple decision trees to improve predictive accuracy and robustness. It is well-suited to handle complex datasets with intricate feature interactions.
  
2. **Gradient Boosting Classifier**: A powerful boosting algorithm that builds models sequentially, each correcting errors from the previous one. It tends to deliver high accuracy, especially on structured data, but at the cost of longer training times.

3. **XGBoost Classifier**: An optimized implementation of gradient boosting, XGBoost is known for its high performance and scalability, often outperforming other algorithms in structured data tasks.

4. **Support Vector Machine (SVM)**: SVM is a robust model for classification tasks, particularly effective in high-dimensional spaces. It aims to find the optimal hyperplane that separates classes with maximum margin.


````python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC

# Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train_scaled, y_train)
rf_pred = rf_classifier.predict(X_test_scaled)

# Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(random_state=42)
gb_classifier.fit(X_train_scaled, y_train)
gb_pred = gb_classifier.predict(X_test_scaled)

# XGBoost Classifier
xgb_classifier = XGBClassifier(random_state=42)
xgb_classifier.fit(X_train_scaled, y_train)
xgb_pred = xgb_classifier.predict(X_test_scaled)

# Support Vector Machine
svm_classifier = SVC(random_state=42)
svm_classifier.fit(X_train_scaled, y_train)
svm_pred = svm_classifier.predict(X_test_scaled)
````

#### **Model Evaluation and Comparison**

To evaluate the performance of these models, I used key metrics such as precision, recall, F1-score, and ROC AUC score. These metrics provide a comprehensive understanding of each model’s ability to accurately predict compliance risks.

````python
from sklearn.metrics import classification_report, roc_auc_score

# Random Forest Evaluation
print("Random Forest - Classification Report:")
print(classification_report(y_test, rf_pred))
print("Random Forest - ROC AUC Score:", roc_auc_score(y_test, rf_pred))

# Gradient Boosting Evaluation
print("Gradient Boosting - Classification Report:")
print(classification_report(y_test, gb_pred))
print("Gradient Boosting - ROC AUC Score:", roc_auc_score(y_test, gb_pred))

# XGBoost Evaluation
print("XGBoost - Classification Report:")
print(classification_report(y_test, xgb_pred))
print("XGBoost - ROC AUC Score:", roc_auc_score(y_test, xgb_pred))

# SVM Evaluation
print("Support Vector Machine - Classification Report:")
print(classification_report(y_test, svm_pred))
print("Support Vector Machine - ROC AUC Score:", roc_auc_score(y_test, svm_pred))
````



**Results:**

- **Random Forest Classifier** achieved the highest accuracy of 96%, with a precision of 0.99 and a recall of 0.93 for the no-distress class. The ROC AUC score was 0.964, indicating strong discriminatory power in distinguishing between distress and no-distress cases.

- **Gradient Boosting Classifier** demonstrated a slightly lower accuracy of 93%, with an ROC AUC score of 0.926. Although it performed well, particularly in identifying distressed companies (recall of 0.97), it was outperformed by the Random Forest in overall accuracy and ROC AUC.

````python
import pandas as pd
import numpy as np

def prepare_data(n_samples=30000):
    np.random.seed(42)

    # Create sample data
    sample_data = pd.DataFrame({
        'Company': np.random.choice(range(1, 11), n_samples),
        'Financial Distress': np.random.uniform(-2, 2, n_samples),
        'x1': np.random.uniform(1000, 100000, n_samples),
        'x2': np.random.uniform(1000, 100000, n_samples),
        'x3': np.random.uniform(1000, 100000, n_samples),
        'loan_amnt': np.random.uniform(1000, 50000, n_samples),
        'person_income': np.random.uniform(20000, 100000, n_samples),
        'loan_status': np.random.choice([0, 1], n_samples),  # 0 for Default, 1 for Current
        'credit_card': np.random.randint(1000000000000000, 9999999999999999, n_samples),
        'transaction_dollar_amount': np.random.uniform(10, 1000, n_samples),
        'Long': np.random.uniform(-180, 180, n_samples),
        'Lat': np.random.uniform(-90, 90, n_samples)
    })

    # Feature engineering
    sample_data['Debt_to_Equity'] = sample_data['x1'] / (sample_data['x2'] + 1e-5)
    sample_data['Lag_Financial_Distress'] = sample_data.groupby('Company')['Financial Distress'].shift(1)
    sample_data.fillna(sample_data.median(), inplace=True)

    return sample_data

sample_data = prepare_data()
````


#### **Compliance Risk Analysis Module**

The compliance risk module addresses the need to assess the potential for loan default (compliance risk). This is done by training a RandomForest model on resampled data to handle class imbalance, ensuring robust predictions.

````python
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE

def compliance_risk_analysis(X, y):
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)

    rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
    rf_model.fit(X_resampled, y_resampled)

    return rf_model

# Prepare data for compliance risk
X_compliance = sample_data.drop(columns=['loan_status', 'credit_card'])
y_compliance = sample_data['loan_status']

# Run Compliance Risk Analysis
rf_model = compliance_risk_analysis(X_compliance, y_compliance)
````

#### **Predictive Analytics Module**

This module focuses on forecasting potential loan defaults using an XGBoost model. The model is trained and tested using a portion of the dataset.

````python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

def predictive_analytics(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    xgb_model = XGBClassifier(n_estimators=200, random_state=42, use_label_encoder=False, eval_metric='logloss')
    xgb_model.fit(X_train, y_train)

    return xgb_model

# Run Predictive Analytics
xgb_model = predictive_analytics(X_compliance, y_compliance)
````



#### **Report Generation Module**

The report generation module automates the creation of a compliance report, which includes predictions on loan status. The report undergoes validation to ensure it meets the required standards before being saved as a CSV file.

````python
def generate_report(sample_data, best_model):
    report_data = sample_data[['Company', 'Financial Distress', 'Debt_to_Equity', 'person_income', 'loan_amnt', 'loan_status', 'credit_card', 'transaction_dollar_amount']]
    report_data['Predicted_Loan_Status'] = best_model.predict(sample_data.drop(columns=['loan_status', 'credit_card']))

    def validate_report(report):
        if report.isnull().values.any():
            raise ValueError("Report contains missing values")

        if not report['Financial Distress'].between(-2, 2).all():
            raise ValueError("Financial Distress values are out of range")

        if not report['Debt_to_Equity'].between(0, 100).all():
            raise ValueError("Debt_to_Equity values are out of range")

        if not report['transaction_dollar_amount'].between(0, 1000).all():
            raise ValueError("Transaction dollar amounts are out of range")

        if not report['loan_status'].isin([0, 1]).all():
            raise ValueError("Loan status contains invalid entries")

        print("Report validation successful")

    validate_report(report_data)
    report_data.to_csv('integrated_regulatory_report.csv', index=False)
    print("Final report generated and saved as 'integrated_regulatory_report.csv'")

# Generate final report
generate_report(sample_data, rf_model)
````



#### **System Integration Module**

This module integrates all the above components into a unified system. It automates the entire process from data preparation to report generation, ensuring that the system is ready for deployment.


````python
def integrate_system():
    sample_data = prepare_data()

    # Split data for compliance risk
    X_compliance = sample_data.drop(columns=['loan_status', 'credit_card'])
    y_compliance = sample_data['loan_status']

    # Run Compliance Risk Analysis
    rf_model = compliance_risk_analysis(X_compliance, y_compliance)

    # Run Predictive Analytics
    predictive_model = predictive_analytics(X_compliance, y_compliance)

    # Generate final report
    generate_report(sample_data, rf_model)

# Run the integrated system
integrate_system()
````

#### **2. IFRS 9 Compliance**

The International Financial Reporting Standard 9 (IFRS 9) requires institutions to implement an Expected Credit Loss (ECL) model for financial instruments, recognizing credit impairments proactively.

**Expected Credit Loss (ECL) Modeling:**

In this example, the system calculates the Expected Credit Loss (ECL) based on the Probability of Default (PD), Loss Given Default (LGD), and Exposure at Default (EAD). The report is generated to ensure compliance with IFRS 9.


````python
import pandas as pd

# Load the credit risk dataset
credit_risk_data = pd.read_csv('/content/credit_risk_dataset.csv')

# Assume loan_int_rate as a proxy for Probability of Default (PD)
credit_risk_data['PD'] = credit_risk_data['loan_int_rate'] / 100

# Assume loan_amnt as the Exposure at Default (EAD)
credit_risk_data['EAD'] = credit_risk_data['loan_amnt']

# Assume a fixed Loss Given Default (LGD) of 45%
credit_risk_data['LGD'] = 0.45

# Calculate Expected Credit Loss (ECL)
credit_risk_data['Expected_Credit_Loss'] = credit_risk_data['PD'] * credit_risk_data['LGD'] * credit_risk_data['EAD']

# Generate IFRS 9 Compliance Report
ifrs_9_report = credit_risk_data[['person_age', 'person_income', 'loan_amnt', 'PD', 'LGD', 'EAD', 'Expected_Credit_Loss']]
ifrs_9_report.to_csv('IFRS_9_Compliance_Report.csv', index=False)

# Display the first few rows of the report
print(ifrs_9_report.head())
````




**Explanation:**
The IFRS 9 Compliance Report details the Expected Credit Loss (ECL) for each company. The ECL is a critical measure used to predict potential credit losses, helping institutions set aside appropriate provisions for impaired loans.
